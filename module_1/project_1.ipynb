# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)101
Варианты ответа:
1. Quentin Tarantino
2. Steven Soderbergh
3. Robert Rodriguez
4. Christopher Nolan
5. Clint Eastwood

```python
answer_ls.append(4)
data['profit'] = data.apply(lambda x: x.revenue - x.budget, axis=1)
maker_films = []
for index,f in data.iterrows():
    for d in f.director.split('|'):
        maker_films += [(f.imdb_id, d)]
directors_df = pd.DataFrame(maker_films, columns=['imdb_id', 'single_director'])
directors_df.describe()
count_y = lambda x: len([ i for i in x if i  > 0])
count_n = lambda x: len([ i for i in x if i <= 0])
data['profit_y'] = data['profit']
data['profit_n'] = data['profit']
filter=['Quentin Tarantino', 'Steven Soderbergh', 'Robert Rodriguez', 'Christopher Nolan', 'Clint Eastwood']
res=directors_df.merge(data, on='imdb_id').groupby('single_director').aggregate({"imdb_id": len, "profit_y": count_y, "profit_n": count_n})
res['p'] = res.profit_y / res.imdb_id
res.sort_values(by='p', ascending=False)
res.reset_index(inplace=True)
res[res.single_director.isin(filter)].sort_values(by='p', ascending=False)
```
